## PROCESSOR DATA RCT DAN RST DARI APKT 

```python
import pandas as pd
import os
from bs4 import BeautifulSoup

# Get the current working directory
dir_path = os.getcwd()

##source == for data from eis apkt, processed == for finished processing file
source_dir_path = os.path.join(dir_path, "source")
processed_dir_path = os.path.join(dir_path, "processed")

#File name penampung data hasil pengolahn/ output file
file_name = "apkt_eis_rpt_rct_combined.csv"

# Get the list of files in the directory
files = os.listdir(source_dir_path)


print("working directory : " + source_dir_path)

if len(files) != 0 :   
    print("jumlah file yang akan di prosess : " + str(len(files)))
    for file in files :
        print("processing file : " + file)
        
        #file directory
        file_path = os.path.join(source_dir_path,file)
            
        #open saved file
        with open(file_path, 'r') as f:
            soup = BeautifulSoup(f, 'html.parser')

        #CHECK TIPE FILE
        df_list = pd.read_html(file_path)

        print(len(df_list[1]))


        ## condition if file uid atau file up3
        if len(df_list[1]) == 1:
            #jika uid proses pakai metode uid
            print("uid")
            parser_data("uid")

        else:
            #jika up3 proses pkaia metode up3
            print("up3")
            parser_data("up3")
        # closed file
        f.close()
        
        print(file)
        
        #move file to processed folder
        os.rename(file_path, os.path.join(processed_dir_path, file))
        
else:
    print("tidak ada data yang akan di prosess di folder source")
        
```

```python
def parser_data(x):
    #kolum data extraksi
    kolum = ['NO','KODE','EQUIPMENT', 'TOTAL', 'JML SDH SELESAI', '% SDH SELESAI', 'JML BLM SELESAI', '% BLM SELESAI', 'RPT TOTAL', 'RPT RATA2', 'RPT MAX', 'RPT MIN', 'RPT > SLA', 'RPT < SLA', 'RCT TOTAL', 'RCT RATA2', 'RCT MAX','RCT MIN', 'RCT > SLA', 'RCT < SLA']
    
   
    
    # Find the specific table in the HTML using its class or id
    table = soup.find('table')

    # Extract the data from the table and store it in a list of lists
    data = []
    for row in table.find_all('tr'):
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        data.append(cols)
        
        
    if x == 'uid' :
        #Exctract date
        
        tanggal = data[7]
        tmp = str.split(tanggal[0])
        print(tmp)
        tanggal = tmp[3]
        print(tanggal)

        data = data[13:]
        df = pd.DataFrame(data, columns=kolum)
        
        #Ubah format tanggal dari indonesia ke english
        format_eng = {"Januari": "January", "Februari" : "February", "Maret": "March", "April": "April","Mei":"May", "Juni" : "June", "Juli": "July", "Agustus": "August", "Oktober": "October", "September":"September","Nopember": "November", "Desember":"December" }
        clean_date = tanggal.split('-')
        print(clean_date)
        tanggal = format_eng[clean_date[1]]+"/"+clean_date[2] 
        print(tanggal)

        #Add date and unit column
        df["unit"] = "UID"
        df["tanggal"] = tanggal
        
        #process data by category
        process_data_unplaned("uid",df)
        process_data_planed("uid",df)
        process_data_bencana("uid",df)
        process_data_pelanggan("uid",df)
        
    else:
        #Exctract date
        unit = df_list[1].iloc[1][1]
        unit = unit[6:]

        print(len(df_list))
        print(unit)
        
        tanggal = data[8]
        tmp = str.split(tanggal[0])
        print(tmp)
        tanggal = tmp[3]
        print(tanggal)

        data = data[13:]
        df = pd.DataFrame(data, columns=kolum)

        #Ubah format tanggal dari indonesia ke english
        format_eng = {"Januari": "January", "Februari" : "February", "Maret": "March", "April": "April","Mei":"May", "Juni" : "June", "Juli": "July", "Agustus": "August", "Oktober": "October", "September":"September","Nopember": "November", "Desember":"December" }
        clean_date = tanggal.split('-')
        print(clean_date)
        tanggal = format_eng[clean_date[1]]+"/"+clean_date[2] 
        print(tanggal)

        #Add date and unit column
        df["unit"] = unit
        df["tanggal"] = tanggal
        
       #process data by category
        process_data_unplaned("up3",df)
        process_data_planed("up3",df)
        process_data_bencana("up3",df)
        process_data_pelanggan("up3",df)
        
```
